#Run experiments to improve the model

1. Get more data: get more data points for the model to train on
2. Make the model larger: Add more layers or more hidden units in each hidden layer
3. Train for longer: Increase the number of epochs



```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = tf.range(-100,100,4)
y = X+10
```

```python
X_train = X[:40]   #80% of the data
y_train = y[:40]

X_test = X[40:]    #20% of the data
y_test = y[40:]
```

## Experiments

1. model_1: 1 layer, 100 epochs
2. model_2: 2 layers, 100 epochs
3. model_3: 2 layers, 500 epochs

```python
model_1 = tf.keras.Sequential([
                               tf.keras.layers.Dense(1)
])

model_1.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["mae"])

model_1.fit(X_train, y_train, epochs = 100)
```

```python
def plot_predictions(train_data,
                     train_labels,
                     test_data,
                     test_labels,
                     predictions):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data, train_labels, c="b", label = "Training data")
  plt.scatter(test_data, test_labels, c = "g", label = "Testing data" )
  plt.scatter(test_data, predictions, c="r", label = "Prediction")
  plt.legend()
```

```python
y_pred_1 = model_1.predict(X_test)
```

```python
plot_predictions(X_train, y_train, X_test, y_test, y_pred_1)
```

```python
A = model_1.evaluate(X_test, y_test)
```

```python
model_2 = tf.keras.Sequential([
                               tf.keras.layers.Dense(100, activation="relu"),
                               tf.keras.layers.Dense(1)
])

model_2.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["mae"])

model_2.fit(X_train, y_train, epochs = 100)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
plot_predictions(X_train, y_train, X_test, y_test, y_pred_2)
```

```python
B = model_2.evaluate(X_test, y_test)
```

```python
model_3 = tf.keras.Sequential([
                               tf.keras.layers.Dense(100, activation="relu"),
                               tf.keras.layers.Dense(1)
])

model_3.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["mae"])

model_3.fit(X_train, y_train, epochs = 500)
```

```python
y_pred_3 = model_3.predict(X_test)
```

```python
plot_predictions(X_train, y_train, X_test, y_test, y_pred_3)
```

```python
C = model_3.evaluate(X_test, y_test)
```

## Comparing results 

```python
import pandas as pd

model_results = [["model_1", A[1]],
                 ["model_2",B[1]],
                 ["model_3", C[1]]]

all_results = pd.DataFrame(model_results, columns = ["model", "mae"])
```
