## Saving a model

There are two main formats to save a model:

1. SavedModel format
2. HDF5 format


```python
import tensorflow as tf
```

```python
X = tf.range(-100,100,4)
y = X+10
```

```python
X_train = X[:40]   #80% of the data
y_train = y[:40]

X_test = X[40:]    #20% of the data
y_test = y[40:]
```

```python
model_3 = tf.keras.Sequential([
                               tf.keras.layers.Dense(100, activation="relu"),
                               tf.keras.layers.Dense(1)
])

model_3.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["mae"])

model_3.fit(X_train, y_train, epochs = 500)
```

```python
#SavedModel format
model_3.save("Best_performing_model_SavedModel_format")
```

```python
#HDF5 format
model_3.save("Best_performing_model_HDF5_format.h5")
```

## Loading a model

```python
#Loading the SavedModel format
loaded_SavedModel_format = tf.keras.models.load_model("Best_performing_model_SavedModel_format")
loaded_SavedModel_format.summary()
```

```python
model_3.summary()
```

```python
# Compare model_3 predictions with the Loaded model predictions
model_3_preds = model_3.predict(X_test)
loaded_SavedModel_format_preds = loaded_SavedModel_format.predict(X_test)
```

```python
model_3_preds == loaded_SavedModel_format_preds
```

```python
# Loading the HDF5 model format
loaded_h5_format = tf.keras.models.load_model("Best_performing_model_HDF5_format.h5")
```

```python
loaded_h5_format_preds = loaded_h5_format.predict(X_test)
```

```python
model_3_preds == loaded_h5_format_preds
```
